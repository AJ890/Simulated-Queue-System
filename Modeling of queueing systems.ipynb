```python
import numpy as np 
import pandas as pd
import schedule
import time
from datetime import datetime, timedelta
from dateutil.parser import parse


def Process_Queue():
    Req = pd.read_csv('Required.csv')
    On_Proc = pd.read_csv('On_Process.csv')
    
    if On_Proc.shape[0] == 0:   
        times = pd.DataFrame({'Arrival': [datetime.now()],
              'Depart': [datetime.now() + timedelta(seconds=60)]})
        rows = Req.loc[:0]
        On_Proc = On_Proc.append(pd.concat([rows,times]
                                ,axis=1,sort=False),
                                 ignore_index=True)
        
        Req.drop(rows.index, inplace=True)
        Req.to_csv('Required.csv',index=False)
        On_Proc.to_csv('On_Process.csv',index=False)
        print(On_Proc)
        
    else:
        print('Number of Queue Processes',On_Proc.shape[0])

def Process_Depart():
    On_Proc = pd.read_csv('On_Process.csv')
    Proc = pd.read_csv('Processed.csv')
    
    if On_Proc.empty:
        print('No Process on Queue')
    
    elif datetime.now()> parse(On_Proc.iloc[0][3]):
        
        times = pd.DataFrame({'Terminated': [datetime.now()]})

        rows = On_Proc[['ID','Type']]

        Proc = pd.concat([pd.concat([rows,times],
                          axis=1),Proc],axis=0,sort=False)

        On_Proc.drop(rows.index, inplace=True)
        Proc.to_csv('Processed.csv',index=False)
        On_Proc.to_csv('On_Process.csv',index=False)
        print(Proc.iloc[0:1])
        
    else:
        print('Process Hasnt Finish',On_Proc.shape[0]) 

def Scheduler():
    Req = pd.read_csv('Required.csv')
    Process_Depart()
    Process_Queue()  

        
schedule.every(30).seconds.do(Scheduler)
Req = pd.read_csv('Required.csv')
while Req.empty==False:
    Req = pd.read_csv('Required.csv')
    schedule.run_pending()        
```
